# 01 — Data audit

Inspect raw Parquet outputs under `data/raw/` after running `helios-pipeline`.


```python
from pathlib import Path
import polars as pl
from helios_alpha.config import load_settings

s = load_settings()
raw = s.data_raw
for p in sorted(raw.rglob("*.parquet")):
    print(p.relative_to(s.repo_root))
    print(pl.read_parquet(p).head(3))
    print()
```
